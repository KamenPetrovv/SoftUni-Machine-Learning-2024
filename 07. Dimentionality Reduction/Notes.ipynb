## Dimentionality Reduction

Reducing the number of rows means we are doing selection. Also called undersampling
- Some algos depend on the number of rows (amount of data) but all algos depend on the number of features/dimensions

Reducing the number of columns is called dimentionality reduction.
- Sometimes more featues dont bring more information. Hence Dimentionaity reduction

## PCA (Principal Component analysis)

When we have many features/params it doesn't mean that they are all useful. What we can do is **Projection**

Projection: is a transformation from a higher-dimentinal dataset to lower-dimentional dataset.
The easiest way to do this transf. is PCA

Example: Shadow.\
Object: 3D, Shadow: 2D

These types of transformations (projections) are not reversible, they dont have an Invertible Matrix. We lose information when we apply it.

The trick is to do it in such a way that we lose the least amount of imformation and simplify the dataset the most.

1. The resulting componenets are ordered by importance: cumul they are 1.

This is not feature selection. We can't remove features. It doesn't mean we have to give less features when we start. It means we have transformed them into lower-dimenitonaliy of a few components.

2. The resulting components/columns are independent

3. The components maximize the described variance

 PCA simply takes points expressed in the standard basis and transforms them into points expressed in an eigenvector basis. In this process of transformation, some dimensions with low variance are discarded and hence the resulting dimensional reduction.

The data is linearly transformed onto a new coordinate system such that the directions (principal components) capturing the largest variation in the data can be easily identified.

PCA doesn't do [Eigen value decomposition](https://en.wikipedia.org/wiki/Eigendecomposition_of_a_matrix) it does [Single value decomposition](https://en.wikipedia.org/wiki/Singular_value_decomposition)

PCA doesn't remove dimentions. It returns the same amount but it orders them by importance. It changes the basis. After that we can decide to remove the last ones that are not as important.

## TRuncatedSVD

It works where PCA doesn't but we lose the order of importance of the components. So we get a reduction in dimentions but maybe not the best one.

PCA doesn't work with sparse matrix

## Linear Discriminant Analysis (LinDA)

Uses PCA under the hood. It transforms the basis in such a way that the CLASSES that we had in the beginning are the most separated. Can be used as dimentionality reduction algo.

## Isometric Mapping (IsoMap)

Seeks a lower-dimensional projection (embedding) which preserves distances between points

Basically does KNN

The distances between the points are preserved (isometric)

## t-SNE (t Distributed Stochastic Neighour Ebedding)

Works only for 2D and 3D, n_components = {2, 3}

Like LinDA preserves the classes, but it doesn't get them. Makes sure the close neighbours remain close.

Stochastic means random

t-SNE gives you a "view" of the data such that it preserves near neighbours. It projects one point at random. Then the same for a second point but it uses T-Distribution, so that the distances are almost the same as the distances as the new view. The clusters are randomly chosen. It is kind of like reverse IsoMap. IsoMap preserves distances, t-SNE preserves neighbours

**Rarely used for Learning, mostly for visualization** (because of limitation of 2D and 3D)\
One of the best ways to visualize embeddings

## UMAP (Uniform Manifild Approximation)

Works almost like a combination of IsoMap and t-SNE, closer to t-SNE.\
Can be used for all dimensions.\
Can be used for feature extraction and learning.

After we apply PCA the accuracy should drop. We are losing some of the imformation, but we are getting a lot faster model.

If we have a model with high bias, DO NOT DO PCA. PCA makes the data in a better structure, but it wont improve it.

The idea is to make it faster. Not more/less accurate. Lose a little bit of prcision, but gain alot of speed.

Careful when used on small datasets. It is best for big datasets
