## [Cover's theorem](https://en.wikipedia.org/wiki/Cover%27s_theorem)

TLDR: If we increase the compleity/dimentionality of the data we allow weaker estimators to do better.

Example: We make from 4 var space to 35 var space with polynomial features and an algorithm like linear regression/LinearSVM can give us way better performance

```Kernal Function - Function that is the Dot Product( ) in higher dimentionality``` Gives us the dot product without making the transformation to higher dimentionality. It is a lot faster in terms of calculations

## Kernal Trick 
### RBF (Radial Basis Function)



The param ```degree=X``` is used only when the kernal is ```kernal='poly'```

With Polynomial Features we know what would be the dimensions of the new dataset.\
With RBF we dont. It does a projection into infinetly long dimensions. 
Think of Taylor series that represent and exponent\
\
We can transform a vector into an infinite one and multiply it by another infinite one that has been transformed and get the result.\
Well RBF kind of does the same but it doesn't have to transform it. Magic.

- With PolyFeat - we have .coef_ 
- With RBF - we dont



```Small gamma => big sigma => big std div => more points can be placed on the same gaussian. (the "hat")```
Assume data is scaled!\
C = L2 regularization\
Gamma - influence of every support vector. How far does each support vector influence spans. Choose numbers smaller than 1.

Small gamma => small overfitting, a lot of regularization\
Big gamma => big overfitting, small regularization

C and Gamma are 2 different things and we optimize them separately.

GridSearch with kernal={poly,rbf} and degree= {1,2,3,4} is a waste of time, cus degree only works with poly

Gamma = "auto" is 1/number of features, this is beause the more features we have the higher the likelyhood that they are correlated.
